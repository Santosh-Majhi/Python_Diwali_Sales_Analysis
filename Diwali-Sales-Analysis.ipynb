```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding="unicode_escape")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"], axis=1, inplace=True)
```

#drop unrelated/blank columns.

```python
df.info()
```

```python
pd.isnull(df).sum()
```

#check for null values...

```python
df.dropna(inplace=True)
```

#drop null values

```python
pd.isnull(df).sum()
```

```python
df["Amount"]= df["Amount"].astype("int")
```

#change data type of amount column.

```python
df["Amount"].dtypes
```

```python
df.columns
```

```python
df.rename(columns={"Marital_Status":"Shaddi"})
```

#renamed marital status to shaddi.

```python
df.describe()
```

```python
df[["Age","Orders","Amount"]].describe()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x="Gender", data=df, color="yellow")
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Gender")
plt.show()
```

```python
sales_gen=df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount")
sns.barplot(x="Gender", y="Amount", data=sales_gen, color="red")
plt.xlabel("Gender")
plt.ylabel("Total Amount")
plt.show()
```

#from the above chart we can see that most of the buyers are females and even the purchasing power of females are greater then men.

```python
plt.figure(figsize=(9,5))
ax=sns.countplot(x="Age Group", data=df, hue="Gender")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

```python
sales_gen=df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount")
sns.barplot(x="Age Group", y="Amount",data=sales_gen,color="green")
plt.show()
```

#from the above chart we have concluded that the most of the buyers are of age group between 26-35 years females.

```python
sales_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(x="State", y="Orders", data=sales_state, color="red")
plt.show()
```

```python
sales_state=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x="State", y="Amount", data=sales_state, color="blue")
plt.show()
```

#from the above charts we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharastra and Karnataka respectively.

```python
plt.figure(figsize=(4,6))
ax=sns.countplot(x="Marital_Status", data=df, color="brown")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sales_state=df.groupby(["Marital_Status","Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
plt.figure(figsize=(6,6))
sns.barplot(x="Marital_Status", y="Amount", data=sales_state, hue="Gender" )
plt.show()
```

#from the above graphs we have concluded that the most of the buyers are married(women) and they have high purchasing power.

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(x="Occupation", data=df, color="cyan")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sales_state=df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
plt.figure(figsize=(20,6))
sns.barplot(x="Occupation", y="Amount", data=sales_state, color="teal" )
plt.show()
```

#from the above chart we can see that most of the buyers are working in IT Sector, Healthcare and Aviation Sector.

```python
plt.figure(figsize=(25,5))
ax=sns.countplot(x="Product_Category", data=df, color="navy")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sales_state=df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
plt.figure(figsize=(20,6))
sns.barplot(x="Product_Category", y="Amount", data=sales_state, color="blue" )
plt.show()
```

#from the above charts we can see that most of the sold products are from Food,Clothing and Electronics Category.

```python
sales_state=df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
plt.figure(figsize=(20,6))
sns.barplot(x="Product_ID", y="Orders", data=sales_state, color="purple" )
plt.show()
```

```python
fig1, ax1=plt.subplots(figsize=(12,7))
df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending=False).plot(kind="bar")
```
